```python
# TAKE  OUT

import numpy as np
import pandas as pd
import pickle
import joblib
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier


from sklearn.metrics import f1_score
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler

from sklearn.pipeline import make_pipeline
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import sklearn

def train_fit(model, param_grid, tranformer, X_train, y_train, X_test, y_test, CV):
    
    pipe = Pipeline([('transformer', transformer), ('model', model)])
    
    # Parameter grid
    grid = GridSearchCV(pipe, param_grid, cv=CV, scoring='f1_micro', refit=True)
    grid.fit(X_train, y_train)
    
    # Fit test data
    y_hat = grid.predict(X_test)
    test_proba = grid.predict_proba(X_test)
    test_score = grid.score(X_test, y_test)

    # Scores
    grid.best_score_
    grid.best_params_
    grid.best_estimator_
    
    # Best model parameters 
    best_params_scores_pred = {
        'train_score':grid.best_score_,
        'test_score': test_score,
        'params':grid.best_params_,
        'estimator':grid.best_estimator_,
        'test_proba': test_proba,
        'y_hat': y_hat
    }
    
    return best_params_scores_pred
```
